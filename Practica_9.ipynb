## <span style="color: var(--vscode-foreground);">Genera una vista para entregar el informe que contenga:</span>

### **4. Las ventas por vendendor**

```python
CREATE VIEW vista_ventas_por_vendedor AS
SELECT 
    VD.ID_Vendedor,
    CONCAT(VD.Nombre, ' ', VD.Apellido) AS Nombre_Vendedor,
    COUNT(*) AS Total_Ventas
FROM ventas V
JOIN vendedor VD ON V.ID_Vendedor = VD.ID_Vendedor
GROUP BY VD.ID_Vendedor, VD.Nombre, VD.Apellido;
```

```python
SELECT 
    VD.ID_Vendedor,
    CONCAT(VD.Nombre, ' ', VD.Apellido) AS Nombre_Vendedor,
    COUNT(*) AS Total_Ventas
FROM ventas V
JOIN vendedor VD ON V.ID_Vendedor = VD.ID_Vendedor
GROUP BY VD.ID_Vendedor, VD.Nombre, VD.Apellido;
```

### **5. Ventas por producto**

```python
CREATE VIEW vista_ventas_por_producto AS
SELECT 
    P.Producto,
    COUNT(*) AS Total_Ventas
FROM ventas V
JOIN producto P ON V.ID_producto = P.ID_Producto
GROUP BY P.Producto;
```

```python
SELECT 
    P.Producto,
    COUNT(*) AS Total_Ventas
FROM ventas V
JOIN producto P ON V.ID_producto = P.ID_Producto
GROUP BY P.Producto;
```

### **6. Resumen de compras por cliente en 2017**

```python
CREATE VIEW resumen_compras_por_cliente_2017 AS
SELECT 
    V.ID_Cliente,
    CONCAT(C.Nombre) AS Nombre_Cliente,
    COUNT(*) AS Total_Compras
FROM ventas V
JOIN clientes C ON V.ID_Cliente = C.ID_Cliente
WHERE YEAR(STR_TO_DATE(Fecha_venta, '%d-%b-%y')) = 2017
GROUP BY V.ID_Cliente, C.Nombre;
```

```python
SELECT 
    V.ID_Cliente,
    CONCAT(C.Nombre) AS Nombre_Cliente,
    COUNT(*) AS Total_Compras
FROM ventas V
JOIN clientes C ON V.ID_Cliente = C.ID_Cliente
WHERE YEAR(STR_TO_DATE(Fecha_venta, '%d-%b-%y')) = 2017
GROUP BY V.ID_Cliente, C.Nombre;
```

# **Having**

## **Tomando en cuenta el ejercicio anterior:**

### **7. Resumen de compras por cliente en 2017 que hayan sido mayores a 150000**

```python
CREATE VIEW resumen_compras_mayores_150k_2017_TRUE AS
SELECT 
    V.ID_Cliente,
    CONCAT(C.Nombre) AS Nombre_Cliente,
    SUM(P.Pago) AS Total_Compras
FROM ventas V
JOIN clientes C ON V.ID_Cliente = C.ID_Cliente
JOIN pagos P ON V.ID_Venta = P.ID_Venta
WHERE YEAR(STR_TO_DATE(V.Fecha_venta, '%d-%b-%y')) = 2017
GROUP BY V.ID_Cliente, C.Nombre
HAVING Total_Compras > 150000;
```

```python
SELECT 
    V.ID_Cliente,
    C.Nombre AS Nombre_Cliente,
    SUM(P.Pago) AS Total_Compras
FROM ventas V
JOIN clientes C ON V.ID_Cliente = C.ID_Cliente
JOIN pagos P ON V.ID_Venta = P.ID_Venta
WHERE YEAR(STR_TO_DATE(V.Fecha_venta, '%d-%b-%y')) = 2017
GROUP BY V.ID_Cliente, C.Nombre
HAVING Total_Compras > 150000;
```

### **8. Proporcione la lista de los clientes morosos**

```python
SELECT 
    V.ID_Cliente,
    C.Nombre AS Nombre_Cliente,
    SUM(V.Venta) - COALESCE(SUM(P.Pago), 0) AS Saldo_Pendiente
FROM ventas V
JOIN clientes C ON V.ID_Cliente = C.ID_Cliente
LEFT JOIN pagos P ON V.ID_Venta = P.ID_Venta
GROUP BY V.ID_Cliente, C.Nombre;
```
